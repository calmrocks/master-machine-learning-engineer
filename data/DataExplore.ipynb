# Overview of Common Data Types

Data comes in diverse forms, each suited to specific applications. Hereâ€™s a brief overview of the main types:

- **Tabular Data**: Structured in rows and columns, like in spreadsheets and databases. Ideal for quantitative analysis and easy querying.  
  *Example*: A table of customer records.

- **JSON Data**: Stores data as key-value pairs in a flexible, hierarchical format. Widely used for APIs and NoSQL databases due to its lightweight, readable structure.  
  *Example*: User profile data in JSON format.

- **Language Data (Text)**: Unstructured, written language data used in NLP tasks. Requires processing techniques to interpret meaning, tone, and context.  
  *Example*: Customer reviews or email text.

- **Image Data**: Visual data made up of pixels. Common in applications like facial recognition and medical imaging.  
  *Example*: A JPEG photo.

- **Video Data**: Sequential images (frames) with audio, capturing motion over time. Essential in fields like surveillance and streaming.  
  *Example*: A tutorial video.

Each data type supports unique applications and analysis methods, enabling versatile insights across industries.


# Overview of Common Data Plotting Methods

Visualizing data is essential for interpreting trends, patterns, and relationships. Here are some popular plotting methods:

- **Line Plot**: Displays data points connected by lines, commonly used to show trends over time.  
  *Example*: Tracking stock prices over months.

- **Bar Chart**: Uses rectangular bars to compare quantities across categories. Great for showing frequency or size differences.  
  *Example*: Sales numbers for different products.

- **Histogram**: Shows the distribution of data by grouping values into bins. Useful for understanding data spread and frequency.  
  *Example*: Age distribution in a survey.

- **Scatter Plot**: Plots individual data points on a grid to reveal relationships between two variables.  
  *Example*: Comparing height and weight.

- **Pie Chart**: Divides data into proportional slices to show part-to-whole relationships. Best for categorical data with few segments.  
  *Example*: Market share by company.

- **Heatmap**: Uses color to represent data values across a matrix. Effective for showing patterns in large datasets.  
  *Example*: Correlation matrix in a dataset.

Each method highlights specific aspects of data, making it easier to communicate insights effectively.


```python
# prompt: write code to load a open source (online) data set. I would like to show the basic data analysis in data. for example, plot of number values,
# box plot, histogram, scatter plot, heat map of words, etc

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (replace with your desired dataset URL)
try:
    data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
except Exception as e:
    print(f"Error loading dataset: {e}")
    print("Please replace 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv' with a valid URL")
    exit()

# Basic data analysis and visualization

# 1. Line Plot (e.g., sepal length over data points)
plt.figure(figsize=(8, 6))
plt.plot(data['sepal_length'], marker='o', linestyle='-')
plt.xlabel('Data Point Index')
plt.ylabel('Sepal Length (cm)')
plt.title('Sepal Length Over Data Points')
plt.show()

# 2. Box Plot (e.g., sepal width across species)
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='sepal_width', data=data)
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Width Distribution by Species')
plt.show()

# 3. Histogram (e.g., distribution of petal length)
plt.figure(figsize=(8, 6))
plt.hist(data['petal_length'], bins=10, edgecolor='black')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Length')
plt.show()

# 4. Scatter Plot (e.g., petal length vs. petal width by species)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=data)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs. Petal Width by Species')
plt.show()

# 5. Bar Chart (e.g., average sepal length for each species)
plt.figure(figsize=(8, 6))
sns.barplot(x='species', y='sepal_length', data=data, ci=None)
plt.xlabel('Species')
plt.ylabel('Average Sepal Length (cm)')
plt.title('Average Sepal Length by Species')
plt.show()

# 6. Pie Chart (proportion of each species in the dataset)
species_counts = data['species'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Each Species')
plt.show()

# 7. Heatmap (correlation matrix for numeric features)
plt.figure(figsize=(10, 8))

# Exclude non-numeric columns
correlation_matrix = data.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu", linewidths=.5)
plt.title('Correlation Matrix of Iris Dataset Features')
plt.show()

# 8. Violin Plot (e.g., distribution of petal length across species)
plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='petal_length', data=data)
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.title('Petal Length Distribution by Species')
plt.show()

```
